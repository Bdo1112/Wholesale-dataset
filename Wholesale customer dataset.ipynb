### Function to Evaluate models

```python
#The function evaluates Precision score, Recall score, and accuracy score for GridSearch
def performance_evaluation(model_name, model, features, labels):
    prediction = model.predict(features)
    
    precision = round(precision_score(labels, prediction, average = 'micro'), 3)
    recall = round(recall_score(labels, prediction, average = 'micro'), 3)
    accuracy = round(accuracy_score(labels, prediction), 3)
    f1 = round((2 * ((precision * recall) / (precision + recall))),3)
    
    print("{} performance are the follwoing".format(model_name))
    print("Precision score is {}".format(precision))
    print("Recall score is {}".format(recall))
    print("Accuracy score is {}".format(accuracy))
    print("F1 score is {}".format(f1))
    
    
```

```python
#The function evaluates Precision score, Recall score, accuracy score and latency for cross_val_score
def performance_evaluation_cvs(model_name, model, feature, labels):
    beg_time = time.time()
    result = cross_val_score(model, feature, labels, cv = kfold)
    
    result_precision = cross_val_score(model, feature, labels, cv = kfold, scoring = 'precision')
    result_recall = cross_val_score(model, feature, labels, cv = kfold, scoring = 'recall')
    Accuracy = result.mean()
    Precision = result_precision.mean()
    Recall = result_recall.mean()
    end_time = time.time()
    f1 = round((2 * ((Precision * Recall) / (Precision + Recall))),3)
    lat_time = end_time - beg_time
    
    print("{} performance are the follwoing".format(model_name))
    print("Precision score is {}".format(round(Precision,3)))
    print("Recall score is {}".format(round(Recall,3)))
    print("Accuracy score is {}".format(round(Precision,3)))
    print("F1 score is {}".format(f1))
    print('Latency on cross_val_score is {}'.format(round(lat_time,3)))
    
```

## Kmean clustering

### Apply KMean Clustering on the scaled dataset with StandardScaler

```python
#Kmeans clustering with number of K between 2 to 15 with for loop.

K = range(2,16)
kSSE_SS = []

for i in K:
    kmeans_SS = KMeans(n_clusters = i)
    kmeans_SS = kmeans_SS.fit(standardScaledData_df)
    SSE_SS = kmeans_SS.inertia_
    kSSE_SS.append(SSE_SS)
    
```

```python
#Saved all SSE values in the list
kSSE_SS
```

```python
#Convert the list into the DataFrame
SSE_score_SS_df = pd.DataFrame({'Clusters' : K, 'WSS' : kSSE_SS})
SSE_score_SS_df
```

```python
#plotting elbow graph with above DataFrame to determine the best number of clusters

plt.figure(figsize = (10,10))
plt.plot(SSE_score_SS_df['Clusters'], SSE_score_SS_df['WSS'])
plt.title("The Elbow Graph for KMean with StandardScaler")
plt.xlabel('The Number of Clusters (K)')
plt.ylabel('WCSS')
plt.xticks(range(2,16,1))
```

### perform XGB Classifier on standardscaled dataset with Gridsearch

```python
#Performed Gridsearch, traning and testing dataset is 6-4, did not split the validation set to ease the process.
XGBC_s_gscv = XGBClassifier(eval_metric='mlogloss')
param_XGBC_s_gscv = {
 'max_depth':range(3,10,2),
 'min_child_weight':range(1,6,2)
}
kfold = KFold(n_splits = 5)

X_train64,X_test64,y_train64, y_test64 = train_test_split(s_X, s_y, 
                                                   test_size = 0.4,
                                                   random_state = 42)

```

```python
#This is to for parameter tuning for XGBClassifier
#Applying gridsear into XGB classifier
gs_start = time.time()
gs_XGBC_s = GridSearchCV(estimator = XGBC_s_gscv, 
 param_grid = param_XGBC_s_gscv, scoring='roc_auc',n_jobs=4, cv= kfold)
gs_end = time.time()

latency_XGBC_mm_gscv = gs_end - gs_start
```

```python
fit_start = time.time()
gs_XGBC_s.fit(X_train64, y_train64.values.ravel())
gs_XGBC_s.cv_results_, gs_XGBC_s.best_params_, gs_XGBC_s.best_score_
fit_end = time.time()
latency_XGBC_mm_fit = fit_end - fit_start
```

```python
#Using performance_evaluation_cvs function to get performance

model_name = 'XGBoost Classifier model with GridSearchCV'
performance_evaluation(model_name, gs_XGBC_s,s_X,s_y)
print('Latency on gridsearch is {}s'.format(round(latency_XGBC_mm_gscv,7)))
print('Latency on traning model is {}s'.format(round(latency_XGBC_mm_fit,3)))
```

The table above shows latency and evaluation score(Precision, Recall, Accuracy and F1) of each model with GridsearchCV method
